## Analytics: Window Functions

```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2

conn = psycopg2.connect("dbname=northwind user=postgres password=password host=localhost port=5433")
```

### Display a table that returns the customers contact names, country and the total number of customers per country.

```python
sql = """
select "ContactName", "Country", 
COUNT(*) OVER(PARTITION by "Country") "CountryCount"
FROM customers
"""

data = sqlio.read_sql_query(sql, conn)
data
```

### Display a table that calculates the running total of unit price for all order.

```python
#RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW can be excluded from script
sql = """
SELECT od."UnitPrice", o."OrderID",o."OrderDate",
sum("UnitPrice") OVER(order by "OrderDate" RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW)
FROM orders o, order_details od
WHERE o."OrderID" = od."OrderID"
"""

data = sqlio.read_sql_query(sql, conn)
data
```
