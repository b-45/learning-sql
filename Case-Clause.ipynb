### Using case clause, write a query that displays the UnitsInStock and it's total supply along with a category of either MINIMUM STOCK LEVEL when units in stock is less than 80, RE-ORDER STOCK LEVEL when units in stock is less than 20, 'MAXIMUM STOCK LEVEL' when units in stock is less than 150

```python
sql = """
SELECT "ProductName", "UnitsInStock",
CASE
    WHEN "UnitsInStock" < 20 THEN 'RE-ORDER STOCK LEVEL'
    WHEN "UnitsInStock" < 80 THEN 'MINIMUM STOCK LEVEL'
    WHEN "UnitsInStock" < 150 THEN 'MAXIMUM STOCK LEVEL'
END as "StockLevels"
FROM products
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```

### Using case clause, write a query that displays a column EmployeeLevel, if HireDate of employee = 1992 then return Senior, if HireDate of employee = 1993 then return Mid-Senior, else if HireDate of employee = 1994 then return Junior

```python
sql = """
SELECT "LastName", "Title", "HireDate", 
CASE
	WHEN date_part('year', "HireDate") = 1992 THEN 'Senior'
	WHEN date_part('year', "HireDate") = 1993 THEN 'Mid-Senior'
	WHEN date_part('year', "HireDate") = 1994 THEN 'Junior'
	END as "EmployeeLevel"
FROM employees
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```
