```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2

conn = psycopg2.connect("dbname=northwind user=postgres password=password host=localhost port=5433")
```

### Display the columns in the Orders & Order_Details table in one result set

```python
# inner join is the cross product of multiple tables and returns a result set that satisfies the condition
sql = """
SELECT *
FROM order_details INNER JOIN orders
ON orders."OrderID" = order_details."OrderID"
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```
