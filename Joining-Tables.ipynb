```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2

conn = psycopg2.connect("dbname=northwind user=postgres password=password host=localhost port=5433")
```

### Display the columns in the Orders & Order_Details table in one result set

```python
# inner join is the cross product of multiple tables and returns a result set that satisfies the condition
sql = """
SELECT *
FROM order_details INNER JOIN orders
ON orders."OrderID" = order_details."OrderID"
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```

### Display table that returns a result a set of  3 relations of Order_Details, Orders, Customers

```python
# inner join is the cross product of multiple tables and returns a result set that satisfies the condition
sql = """
SELECT *
FROM order_details INNER JOIN orders
ON orders."OrderID" = order_details."OrderID"
INNER JOIN customers
ON orders."CustomerID" = customers."CustomerID"
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```

### Display a table whose return set combines phone numbers from customer & suppliers table

```python
# union join stacks tables 

sql = """
SELECT "Phone" 
FROM customers
UNION
SELECT "Phone"
FROM suppliers
"""

data = sqlio.read_sql_query(sql, conn)
data.head(10)
```

```python
# switch database
conn2 = psycopg2.connect("dbname=mastersql user=postgres password=password host=localhost port=5433")
```

### Display all departments in department table and employees table giving prefernce to department table

```python
# left join returns all matches from left table whether or not they have matches in the table in the LEFT JOIN clause
sql = """
SELECT departments.department, employees.department
FROM departments LEFT JOIN employees
ON departments.department = employees.department
"""

data = sqlio.read_sql_query(sql, conn2)
data.head(10)
```

```python
# right join returns all matches from left table whether or not they have matches in the table in the RIGHT JOIN clause
sql = """
SELECT departments.department, employees.department
FROM departments RIGHT JOIN employees
ON departments.department = employees.department
"""

data = sqlio.read_sql_query(sql, conn2)
data.head(10)
```
